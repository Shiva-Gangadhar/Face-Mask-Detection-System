```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib
```

```python
model = load_model("face_mask.h5")  # or wherever you've saved it

# Load OpenCV's pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

IMG_SIZE = 128  # your model input size

# Start webcam
cap = cv2.VideoCapture(0)  # 0 = default camera

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        # Extract face region
        face_img = frame[y:y+h, x:x+w]

        # Preprocess the face for your model
        face_img_resized = cv2.resize(face_img, (IMG_SIZE, IMG_SIZE))
        face_img_normalized = face_img_resized / 255
        face_input = np.reshape(face_img_normalized, (1, IMG_SIZE, IMG_SIZE, 3))

        # Predict mask / no-mask
        prediction = model.predict(face_input)
        label_index = np.argmax(prediction)

        label = "Mask " if label_index == 1 else "No Mask "
        color = (0, 255, 0) if label_index == 1 else (0, 0, 255)

        # Draw bounding box and label
        cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
        cv2.putText(frame, label, (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

    # Show live feed
    cv2.imshow("Real-Time Mask Detection", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
